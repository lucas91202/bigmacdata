```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.DataFrame(pd.read_csv('data/big_mac_source_data_v2.csv'))
df = df.drop(columns=['iso_a3', 'GDP_dollar', 'GDP_local'])
df = df.query("name =='Australia' | name =='Britain' | name =='France' | name =='New Zealand' | name =='United States'")
df['usd_price'] = df.local_price*df.dollar_ex

def func(mean, median):
    x = ["Australia", "Britain", "France", "New Zealand", "United States"]
    y = []
    if mean == True:
        for z in x:
            a = df.loc[df["name"] == z]
            a_mean = a["usd_price"].mean()
            y.append(a_mean)
        plt.bar(x, y, 0.5)
        plt.show()
    y.clear()
    if median == True:
        for z in x:
            a = df.loc[df["name"] == z]
            a_median = a["usd_price"].median()
            y.append(a_median)
        plt.bar(x, y, 0.5)
        plt.show()

func(1, 0) #First position is mean, second is median
```
